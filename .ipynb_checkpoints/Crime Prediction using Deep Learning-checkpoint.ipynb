```python
# Let's visualise the dataset but before it we need to implement some helper functions
def visualise_df(df, title):
    """
    Helps in plotting the df with the title
    Inputs:
        df-> a dataframe or a dictionary
        title-> title of the plot
    """
    # we need to first count the total crime assigned to a particular key
    # for e.g. WARRANTS -> 450, SUICIDE->300, here Counter will compute 450, 300 based on key WARRANTS and SUICIDE
    mapping = dict(Counter(df))
    labels, values = mapping.keys(), mapping.values()
    fig, ax = plt.subplots(figsize=(12,9))
    plt.title(title)
    plt.barh(range(len(labels)), list(values), color="#33D5FF")
    plt.yticks(range(len(labels)), labels)
    plt.plot()

def visualise_based_on_descript(descript):
    newDf = df[df.Descript == descript]
    if(len(newDf) == 0):
        print("No incident found for: {}".format(descript))
    else:
        print("Total {0} incident found, for {1} DESCRIPT".format(len(newDf), descript))
        # now call visualise_df
        visualise_df(newDf["Resolution"], "Resolutions")  # will tell us how many were actually arrested and prosecuted
        visualise_df(newDf["DayOfWeek"], "Day of the Week")
```

```python
visualise_df(df.Category, "Categories of Crimes")
```

```python
# Inference from the above figure
# we're dealing with LARCENCY/THEFT which seems to be most common crime
# as you can see second position is been secured by OTHER OFFENSES, what are those basically? well we can get a better insight
# if we look into the DESCRIPT of those
visualise_df(dict(Counter(df[df.Category=="OTHER OFFENSES"].Descript).most_common(20)), "Top 20 Other OFFENSES")
```

```python
# so we've DRIVER LICENSE, SUSPENDED OR REVOKED that tops the OTHER OFFENSES
# how about if we look into the DESCRIPT of the top 30 crimes and not only the OTHER OFFENSES
visualise_df(dict(Counter(df.Descript).most_common(30)), "Top 30 Crimes")
```

<p> What we see here is that GRAND THEFT FROM LOCKED AUTO is the most pepetrated crime of all crime followed by LOST PROPERTY and then BRIBERY</p>
